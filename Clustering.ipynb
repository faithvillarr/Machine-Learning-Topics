# K-Means Clustering

We'll be analyzing the most popular dataset of all time: the Iris dataset! We'll be using K-Means clutsering to identify what iris are which. 

```python
# Import necessary libraries
from sklearn import datasets
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Load dataset into numpy arrays
iris = datasets.load_iris()

data  = iris.data
target = iris.target
target_names = iris.target_names
feature_names = iris.feature_names

data = np.array(iris.data)
target = np.array(iris.target)
```

### Plotting Simple KMeans
Since we know there are three target types, we can set there to be three clusters. Lets start with the simplest use of scikit-learn's KMeans() function. 

First we will use random initialization. 

```python
# Apply K-Means Clustering
kmeans_rand = KMeans(n_clusters=3, init='random')
kmeans_rand.fit(data)

centroids = kmeans_rand.cluster_centers_
labels = kmeans_rand.labels_

# Create 3d plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(2, 2, 1, projection="3d", elev=48, azim=134)
kmeans_rand.fit(data)
ax.scatter(data[:, 3], data[:, 0], data[:, 2], c=labels.astype(float), edgecolor="k")

# Plotting centeroids
centroids_x = centroids[:, 3]
centroids_y = centroids[:, 0]
centroids_z = centroids[:, 2]
ax.scatter(centroids_x, centroids_y, centroids_z, c='blue', s=200, marker='^', label='Centroids')


ax.set_xlabel("Petal width")
ax.set_ylabel("Sepal length")
ax.set_zlabel("Petal length")
ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
ax.zaxis.set_ticklabels([])
ax.set_title("3 Clusters. Random Initialization.")

plt.show()
```

### 'Greedy' K-Means
Scikit-learn uses k-means++ initililzation as a defualt. Lets see the results below. 

```python
# Apply K-Means Clustering
kmeans_rand = KMeans(n_clusters=3)
kmeans_rand.fit(data)

centroids = kmeans_rand.cluster_centers_
labels = kmeans_rand.labels_

# Create 3d plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(2, 2, 1, projection="3d", elev=48, azim=134)
kmeans_rand.fit(data)
ax.scatter(data[:, 3], data[:, 0], data[:, 2], c=labels.astype(float), edgecolor="k")

# Plotting centeroids
centroids_x = centroids[:, 3]
centroids_y = centroids[:, 0]
centroids_z = centroids[:, 2]
ax.scatter(centroids_x, centroids_y, centroids_z, c='blue', s=200, marker='^', label='Centroids')


ax.set_xlabel("Petal width")
ax.set_ylabel("Sepal length")
ax.set_zlabel("Petal length")
ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
ax.zaxis.set_ticklabels([])
ax.set_title("3 Clusters. Random Initialization.")

plt.show()
```
